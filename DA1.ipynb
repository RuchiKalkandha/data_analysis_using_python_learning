```python
import pandas as pd

# Create a sample DataFrame
data = {
    'A': [1, 2, 3, 4],
    'B': [5, 6, 7, 8],
    'C': [9, 10, 11, 12]
}
df1 = pd.DataFrame(data)

# Using .loc to select data by label
# .loc is label-based. It selects rows and columns by their labels (names/indices).
# Syntax: df.loc[row_label, column_label]
print(df1.loc[1, 'B'])  # Outputs: 6 (row with index label 1, column 'B')

# Using .iloc to select data by position
# .iloc is integer position-based. It selects rows and columns by their integer positions.
# Syntax: df.iloc[row_position, column_position]
print(df1.iloc[1, 1])  # Outputs: 6 (second row, second column)

# Important Points:
# 1. .loc uses labels (names/indices) for selection.
# 2. .iloc uses integer-based indexing for selection.
# 3. .loc can be used with boolean arrays for conditional selection.
# 4. .iloc is best for selecting data by position when you don't know the exact label.
```
